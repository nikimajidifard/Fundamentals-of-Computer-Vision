```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
import os
```

<h3><font color = "#079e05"> Read input images from 1 to 8 and show them in a row together </font></h3>

```python
imgs=[]
for i in range(1,9):
    imgs.append(cv2.imread(f'images/Q2/{i}.jpg' ))
    imgs[i-1] = cv2.resize(imgs[i-1] ,(0,0) ,fx=0.5 ,fy=0.5)

    j = 0
fig, (ax1, ax2 , ax3,ax4,ax5,ax6,ax7,ax8) = plt.subplots(1, 8,figsize =(18, 18) ,
                                    sharex=True, sharey=True)
tuples = (ax1, ax2 , ax3,ax4,ax5,ax6,ax7,ax8)
for i in range(8):
        ax = tuples[i]
        ax.imshow(imgs[i])
        ax.axis('off')
    
    
   

plt.show()
```

<h3><font color = "#079e05"> Initialize OpenCV's image sticher object and then perform the image stitching on input images </font></h3>

```python
stitcher = cv2.Stitcher_create(0)
status, stitched = stitcher.stitch(imgs)
```

<h3><font color = "#079e05"> Show the result </font></h3>

```python

plt.imshow(stitched)
plt.axis('off')
```
